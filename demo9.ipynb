### polymorphism

#### Duck typing

```python
class Pycharm:
    def execute(self):
        print("Compiling")
        print("Running")

class myedit:
    def execute(self):
        print("Compiling")
        print("Running")
        print("Joss editor")


class Laptop:
    def code(self, ide):
        ide.execute()

ide = myedit()

lap1 = Laptop()

lap1.code(ide)
```
