### Types of Variables in python

```python
class Car:
    def __init__(self):
        self.mil = 10
        self.com = "BMW" #this variables are instace variable because based on my preferences i need top change their values 

c1 = Car()
c2 = Car()

print(c1.com, c1.mil)
print(c2.com, c2.mil)
```

```python
class Car:
    def __init__(self):
        self.mil = 10
        self.com = "BMW" #this variables are instace variable because based on my preferences i need top change their values 

c1 = Car()
c2 = Car()

c1.mil = 8

print(c1.com, c1.mil)
print(c2.com, c2.mil)
```

```python
class Car:
    #if i define a variable outside of __init__ it becomes a class variable 
    #what is class variable?
    #it basically works like a global variable; if i chnage it, it will effect to each and every object 
    wheels = 4


    def __init__(self):
        self.mil = 10
        self.com = "BMW" #this variables are instace variable because based on my preferences i need top change their values 

c1 = Car()
c2 = Car()

c1.mil = 8

print(c1.com, c1.mil, c1.wheels)
print(c2.com, c2.mil, c2.wheels)
```

```python
# another important thing 
# the variable inside the __init__ are instance namespace
# so if we want to change it, we use object name to change it example c1.mil = 8
# for the class variables, if we want to change it we need to mention class name 
# example : if we need to change wheel vaiable we need to chgange it like Car.wheels = 5
```
