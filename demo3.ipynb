```python
class Computer():
    def __init__(self):
        self.name = "Abir"
        self.age = "28"

c1 = Computer()
c2 = Computer()

c1.name = "Mebin"
c1.age = "25"

print(c1.name)
print(c2.name)
```

```python
class Computer():
    def __init__(self):
        self.name = "Abir"
        self.age = "28"

    def update(self):
        self.age = 30

c1 = Computer()
c2 = Computer()

c1.name = "Mebin"
c1.age = "25"

c1.update()

print(c1.age)
print(c2.name)
```

```python
class Computer():
    def __init__(self):
        self.name = "Abir"
        self.age = "28"

    def compare(self, other):
        if self.age == other.age:
            return True
        else:
            return False

c1 = Computer()
c2 = Computer()

c1.name = "Mebin"
c1.age = 30
c1.age = "25"

if c1.compare(c2):
    print("They are same")
else:
    print("They are different")




print(c1.age)
print(c2.name)
```
