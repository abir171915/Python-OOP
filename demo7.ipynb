### Inheritance

```python
class A:
    def feature1(self):
        print("feature 1 is working")
    def feature2(self):
        print("feature 2 is working")
a1 = A()

a1.feature1()
a1.feature2()
```

```python
class A:
    def feature1(self):
        print("feature 1 is working")
    def feature2(self):
        print("feature 2 is working")


class B:
    def feature3(self):
        print("feature 3 is working")
    def feature4(self):
        print("feature 4 is working")        


a1 = A()
b1 = B()

a1.feature1()
a1.feature2()
b1.feature3()
#b1 object can  not access feature 1 or 2 as they are from class A and b1 object is from class B
#but in many cases I might need to access the method from other classes 
#that's where inheritance comes
```

##### single level inheritance

```python
class A:
    def feature1(self):
        print("feature 1 is working")
    def feature2(self):
        print("feature 2 is working")


class B(A):
    def feature3(self):
        print("feature 3 is working")
    def feature4(self):
        print("feature 4 is working")        
#here class B is inheriting class A 
#means class B is the child class of class A 
#here A is the super class/ parent class 
#here B is the sub class/ child class

b1 = B()

b1.feature1()
```

#### multi level inheritance

```python
class A:
    def feature1(self):
        print("feature 1 is working")
    def feature2(self):
        print("feature 2 is working")


class B(A):
    def feature3(self):
        print("feature 3 is working")
    def feature4(self):
        print("feature 4 is working")   

class C(B):
    def feature5(self):
        print("feature 5 is working")

c1 = C()

c1.feature1()
    
```

#### multi level inheritance

```python
class A:
    def feature1(self):
        print("feature 1 is working")
    def feature2(self):
        print("feature 2 is working")


class B():
    def feature3(self):
        print("feature 3 is working")
    def feature4(self):
        print("feature 4 is working")   

class C(A,B):
    def feature5(self):
        print("feature 5 is working")

c1= C()

c1.feature1()

#here class is inherating both classes A and B
```
