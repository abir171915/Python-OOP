```python
class Computer:

    def __init__(self, cpu, ram):
        self.cpu = cpu 
        self.ram = ram

    def config(self):
        print("config is", self.cpu, self.ram)

com1 = Computer("i3", 16) # Constructor
com2 = Computer("Ryzen 3", 8) # Constructor

com1.config()
com2.config()
```
