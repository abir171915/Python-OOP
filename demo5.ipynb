### Types of methods in python

```python
# we have 3 types of methods 
# 1. Class methods
# 2. Instance methods 
# 3. Static methods  
```

```python
class Student:

    school = "UOB"


    def __init__(self, m1, m2, m3):
        self.m1 = m1
        self.m2 = m2
        self.m3 = m3

    def avg(self):
        return (self.m1 + self.m2 + self.m3)/3
    
    # the method avg is an instance method that's why it worked with object

    def get_m1(self):   # this is a method that only fetch the value that's why it is called accessors
        return self.m1
    
    def set_m1(self, value): #this is a method that change the value that's why it is called mutators 
        self.m1 = value

    # both accessors and mutators are types of instance methods because they all work with objects


    # if i want to work with class variable, I need a class method which is bellow
    @classmethod #decorator
    def get_school_name(cls):
        return cls.school
    
    #this method is nothing to do with class variables or instance variables
    #this is what static methods are 
    @staticmethod
    def info():
        print("This is student class")
        
s1 = Student(34,85,98)
s2 = Student(44,77,22)


print(s1.avg())
print(s2.avg())
print(Student.get_school_name())
Student.info()
```
