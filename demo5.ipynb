### Types of methods in python

```python
# we have 3 types of methods 
# 1. Class methods
# 2. Instance methods 
# 3. Static methods  
```

```python
class Student:

    school = "UOB"
    def __init__(self, m1, m2, m3):
        self.m1 = m1
        self.m2 = m2
        self.m3 = m3

    def avg(self):
        return (self.m1 + self.m2 + self.m3)/3
    
    # the method avg is an instance method that's why it worked with object

s1 = Student(34,85,98)
s2 = Student(44,77,22)


print(s1.avg)
print(s2.avg())
```
